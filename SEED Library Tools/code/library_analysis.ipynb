```python
import pandas as pd
import cfg
```

```python


def calculate_skew():
    count_matrix = pd.read_csv(cfg.DATA_DIR+'mapped_counts.csv',index_col = 0)
    count_matrix['depth_flags'] = ''
    count_matrix['sequence_freq'] = ''
    total_reads = count_matrix.sum()['reads']
    total_sequences = len(count_matrix)
    expected_freq = total_sequences/total_reads

    depth_flag_cutoff = 100

    for sequence in count_matrix.index:
        sequence_reads = count_matrix.loc[sequence, 'reads']
        if sequence_reads < depth_flag_cutoff: count_matrix[sequence, 'depth_flags'] = 'FLAG'
        sequence_freq = sequence_reads/total_reads
        count_matrix[sequence, 'sequence_freq'] = sequence_freq
    print(count_matrix)


calculate_skew()
```
